IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

IMPORTING THE DATASET

```python
dataset = pd.read_csv('C:/Users/ViVian/Desktop/VIVIAN/MLworks/regression/Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

TRAINING THE LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
linear_Regressor = LinearRegression()
linear_Regressor.fit(x,y)
```

TRAINING THE POLYNOMIAL REGRESSION MODEL

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor = PolynomialFeatures(degree=4)
x_poly = poly_regressor.fit_transform(x)
linear_Regressor2 = LinearRegression()
linear_Regressor2.fit(x_poly,y) 
```

VISUALIZING THE LINEAR REGRESSION MODEL

```python
plt.scatter(x,y,color='red')
plt.plot(x, linear_Regressor.predict(x),color='blue')
plt.title('TRUTH VS BLUFF(linear regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

VISUALIZING THE POLYNOMIAL REGRESSION MODEL

```python
plt.scatter(x,y,color='red')
plt.plot(x,linear_Regressor2.predict(x_poly),color='blue')
plt.title('TRUTH VS BLUFF(polynomial regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

PREDICTING A NEW RESULT WITH LINEAR REGRESSION

```python
print(linear_Regressor.predict([[6.5]]))
#ans = 330378.78787879 --> ist way off 
```

PREDICTING A NEW RESULT WITH POLYNOMIAL REGRESSION

```python
linear_Regressor2.predict(poly_regressor.fit_transform([[6.5]]))
```
