IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

IMPORTING THE DATASET

```python
dataset = pd.read_csv('C:/Users/ViVian/Desktop/VIVIAN/MLworks/regression/Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
print(x)
print(y)
```

TRAINING THE RANDOM FOREST REGRESSION MODEL ON THE WHOLE DATASET

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10,random_state=0)
regressor.fit(x,y)
```

PREDICTING THE RESULT 

```python
regressor.predict([[6.5]])
```

VISUALIZING THE RANDOM FOREST REGRESSION RESULT

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='red')
plt.plot(x_grid,regressor.predict(x_grid),color = 'blue')
plt.title('TRUTH VS BLUFF(RANDOM FOREST REGRESSION)')
plt.xlabel('position salary')
plt.ylabel('salary')
plt.show()
```
