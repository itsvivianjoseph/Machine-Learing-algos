IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

IMPORTING THE DATASET

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values
```

USING DENDOGRAMS FOR FINDING THE OPTIMAL NUMBER OF CLUSTERS

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(x, method ='ward'))
plt.title('dendogram')
plt.xlabel('observation points')
plt.ylabel('euclidean distance')
plt.show()
```

TRAINING THE HEIRARCHIAL CLUSTERING MODEL ON THE DATASET

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(x)
print(y_hc)
```

VISUALIZING THE CLUSTERS

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=80,c='red',label='cluster1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=80,c='blue',label='cluster2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=80,c='yellow',label='cluster3')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=80,c='grey',label='cluster4')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=80,c='cyan',label='cluster5')
plt.title('CLUSTERS OF CUSTOMERS')
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING SCORE')
plt.show()
```
